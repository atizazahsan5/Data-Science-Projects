# **Rock vs Mine Prediction (A Logistic Regression Problem)**

The project aims to develop a machine learning model capable of accurately differentiate between underwater rocks and explosive mines based on SONAR return data.

```python
# Loading the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #for evaluating the model
```

```python
df = pd.read_csv("/content/sonar data - sonar data.csv")
```

```python
df.tail()
```

```python
df.shape
```

```python
df.describe().transpose()
```

```python
df["61"].value_counts()
```

```python
X = df.drop(columns = "61", axis = 1)
Y = df["61"]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state=1)
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
logit = model.fit(X_train, Y_train)
print(logit)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on testing data : ", testing_data_accuracy)
```
